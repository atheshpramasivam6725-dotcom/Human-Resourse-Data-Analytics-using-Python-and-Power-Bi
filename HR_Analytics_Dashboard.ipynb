```python
# ======================================================
#  HR Analytics Dashboard (Employee Attrition Analysis)
# ======================================================

#  Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ======================================================
#  Step 2: Load Dataset
# ======================================================
data = pd.read_csv("/content/Cleaned_HR_Data.csv")

print("===== Dataset Information =====")
print(data.info())

print("\n===== Checking Missing Values =====")
print(data.isnull().sum())

# Drop missing values if any
data.dropna(inplace=True)

# ======================================================
#  Step 3: Basic Statistics
# ======================================================
print("\n===== Statistical Summary =====")
print(data.describe())

# ======================================================
#  Step 4: Exploratory Data Analysis (EDA)
# ======================================================

# Overall Attrition Count
plt.figure(figsize=(6,4))
sns.countplot(data=data,hue='Attrition', x='Attrition', palette='Set2')
plt.title('Overall Employee Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()

# Attrition by Department
plt.figure(figsize=(7,5))
sns.countplot(data=data, x='Department', hue='Attrition', palette='coolwarm' )
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.show()

# Attrition by Age Distribution
plt.figure(figsize=(7,5))
sns.histplot(data=data, x='Age', hue='Attrition', multiple='stack', palette='Set1')
plt.title('Attrition by Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()

# Monthly Income vs Attrition
plt.figure(figsize=(6,4))
sns.boxplot(data=data, hue='Attrition', x='Attrition', y='MonthlyIncome', palette='pastel')
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()

# Correlation Heatmap (for numeric columns only)
plt.figure(figsize=(10,8))
sns.heatmap(data.select_dtypes(include='number').corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

# ======================================================
#  Step 5: Export Cleaned Data for Power BI
# ======================================================
data.to_csv("Cleaned_HR_Data.csv", index=False)
print("\n Cleaned data exported successfully for Power BI dashboard.")

```
